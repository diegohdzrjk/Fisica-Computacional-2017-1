```python
from numpy import *
from matplotlib.pyplot import *
%matplotlib inline
```

```python
def euler(funcion, p_ini, tf, h=0.01):
    tiempos = arange(p_ini[0], tf, h)
    x = [0]*len(tiempos)
    x[0] = p_ini[1]
    for i in range(0, len(tiempos)-1):
        x[i+1] = x[i]+h*funcion(x[i], tiempos[i])
    solucion = [tiempos, x]
    return solucion
```

Si queremos resolver
$$f'(x) = f(x)$$
$$f(0)=1$$
que sabemos que la solucion es $f(x) = e^{x}$ ponemos:

```python
def func(x, t):
    return x

sol = euler(func, [0, 1], 1, 0.01)
plot(sol[0], sol[1], sol[0], exp(sol[0]));
```

Si ahora queremos resolver
$$ f''(x)+f(x)=0$$ 
$$ f(0) = 0$$$$ f'(0) = 1$$
que sabemos que la solucion es f(x) = sin(x)

```python
def func_segundo(x, t):
    return array([x[1],-x[0]])

sol = euler(func_segundo, [0, array([0,1])], 20, 0.01)


y = zeros(len(sol[0]))
v = zeros(len(sol[0]))


for i in range(len(sol[0])):
    y[i] = sol[1][i][0]
    v[i] = sol[1][i][1]

plot(sol[0], y, sol[0], sin(sol[0]));
```

Si mostramos el espacio fase tenemos

```python
plot(y, v, sin(sol[0]), cos(sol[0]));
```
